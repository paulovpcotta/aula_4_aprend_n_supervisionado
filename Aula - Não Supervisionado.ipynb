```python
# Imports do que vamos utilizar
import numpy as np
from sklearn.datasets.samples_generator import make_blobs
from sklearn.mixture import GaussianMixture
from matplotlib import pyplot as plt
import seaborn as sns

# Dataset
sns.set()
```

```python
# Gráficos que devemos entender
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
plt.scatter(X[:,0], X[:,1])
```

#### O que vimos?

Que a distribuição é mais simples para encontrar os clusters.

```python
# situação não tão linear
n_components = np.arange(1, 21)
models = [GaussianMixture(n, covariance_type='full', random_state=0).fit(X) for n in n_components]
plt.plot(n_components, [m.bic(X) for m in models], label='BIC')
plt.plot(n_components, [m.aic(X) for m in models], label='AIC')
plt.legend(loc='best')
plt.xlabel('n_components');
```

### Nós treinamos nosso modelo usando o número ideal de clusters (neste caso, 4).

```python
# Vamos de GMM agora
gmm = GaussianMixture(n_components=4)
gmm.fit(X)
```

### Utilizamos o método de **predict** para obter uma lista de pontos e seus respectivos clusters.

```python
labels = gmm.predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis');
```

#### Diferentemente de k-means, os modelos de mistura gaussiana são responsáveis ​​pela variação e retornam a probabilidade de que um ponto de dados pertença a cada um dos K ​​clusters.
